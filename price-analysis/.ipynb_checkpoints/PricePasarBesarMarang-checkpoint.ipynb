# Read the data and parse accordingly

```python
import pandas as pd

URL_DATA = 'https://storage.data.gov.my/pricecatcher/lookup_premise.parquet'

df_location = pd.read_parquet(URL_DATA)
if 'date' in df_location.columns: df_location['date'] = pd.to_datetime(df_location['date'])

df_location.head()
```

```python
filter_location = df_location["premise"] == "PASAR BESAR MARANG"
df_temp = df_location[filter_location]
df_temp.head()
```

```python
# pass link and read to the instantiated dataframe
def read_todf(link) :

    global df
    
    df_read = pd.read_parquet(link)
    if 'date' in df_read.columns: df_read['date'] = pd.to_datetime(df_read['date'])

    filter_marang = df_read["premise_code"] == 7997
    df_temp = df_read[filter_marang]

    for i in range(len(df_temp["date"])) :
        df.loc[len(df["Date"])] = df_temp.iloc[i].tolist()

for i in URL_LIST :
    read_todf(i)

df.head()
```

# Explore data

checking harga ayam

```python
harga_ayam_filter = df["Item Code"] == 1
harga_ayam = df[harga_ayam_filter]

harga_ayam.head(40)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

harga_ayam["Date Ordinal"] = harga_ayam["Date"].apply(lambda x : x.toordinal())

plt.figure(figsize = (30, 20))
sns.regplot(x = "Date Ordinal", y = "Price", data = harga_ayam)
plt.gca().set_xticklabels([])

plt.xlabel("Date progression from early 2022 to end 2023 (day)")
plt.ylabel("Price (RM)")
```

```python
# cleanup for easy calculation for the next part
harga_ayam.drop(columns = ["Date Ordinal"], inplace = True)
```

Group and calculate mean per months

```python
monthly_mean = harga_ayam.groupby(pd.Grouper(key='Date', freq='ME')).mean()
monthly_mean = monthly_mean.reset_index()

harga_ayam['YearMonth'] = harga_ayam['Date'].dt.to_period('M')
monthly_mean = harga_ayam.groupby('YearMonth').mean()
monthly_mean.drop(columns = ["Date"], inplace = True)
```

```python
monthly_mean = monthly_mean.reset_index()
monthly_mean.head(10)
```

```python
monthly_mean.columns
```

```python
# introduce ordinal date and plot regression
monthly_mean["Date Ordinal"] = monthly_mean["YearMonth"].apply(lambda x : x.to_timestamp().toordinal())

plt.figure(figsize = (30, 20))
sns.regplot(x = "Date Ordinal", y = "Price", data = monthly_mean, label = "Price regression")
sns.lineplot(x = "Date Ordinal", y = "Price", data = monthly_mean, label = "Price average (per month)")
plt.gca().set_xticklabels([])

plt.xlabel("Date progression from early 2022 to end 2023 (month)")
plt.ylabel("Price (RM)")
plt.legend()
```
