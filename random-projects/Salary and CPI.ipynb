***Import required library and dataset***

```python
import seaborn as sns
import numpy as np
import matplotlib as plt
import pandas as pd
```

```python
URL_DATA = 'https://storage.googleapis.com/dosm-public-economy/cpi_headline.parquet'
URL_DATA2 = 'https://storage.googleapis.com/dosm-public-economy/salaries_industry_sex.parquet'

df1 = pd.read_parquet(URL_DATA)
if 'date' in df1.columns: df1['date'] = pd.to_datetime(df1['date'])

df2 = pd.read_parquet(URL_DATA2)
if 'date' in df2.columns: df2['date'] = pd.to_datetime(df2['date'])
```
